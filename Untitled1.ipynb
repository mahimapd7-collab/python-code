```python
# Creating dictionaries in Python# 1. Direct creation using curly braces
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}
print("Direct dictionary:", person)

# 2. Using the dict() constructor
person2 = dict(name="Bob", age=30, city="London")
print("Using dict() constructor:", person2)

# 3. Creating an empty dictionary and adding items
person3 = {}
person3["name"] = "Charlie"
person3["age"] = 28
person3["city"] = "Paris"
print("Empty then add:", person3)

# 4. From a list of tuples (key-value pairs)
pairs = [("name", "David"), ("age", 35), ("city", "Berlin")]
person4 = dict(pairs)
print("From list of tuples:", person4)

# 5. Dictionary comprehension
squares = {x: x**2 for x in range(1, 6)}
print("Dictionary comprehension:", squares)

# 6. Safe creation with validation
def create_dict_safe(keys, values):
    """
    Creates a dictionary from keys and values with validation.
    Ensures keys are unique and immutable.
    """
    if not isinstance(keys, (list, tuple)) or not isinstance(values, (list, tuple)):
        raise TypeError("Keys and values must be lists or tuples.")
    if len(keys) != len(values):
        raise ValueError("Keys and values must have the same length.")
    if len(set(keys)) != len(keys):
        raise ValueError("Keys must be unique.")
    return dict(zip(keys, values))

try:
    keys = ["id", "name", "age"]
    values = [101, "Eve", 29]
    safe_dict = create_dict_safe(keys, values)
    print("Safe dictionary:", safe_dict)
except (TypeError, ValueError) as e:
    print("Error creating dictionary:", e)
```

```python

data = {
    'Roll_No': [101, 102, 103, 104, 105],
    'Name': ['Aman', 'Bhavna', 'Chetan', 'Divya', 'Esha'],
    'Branch': ['CSE', 'ECE', 'CSE', 'ME', 'EE'],
    'CGPA': [8.5, 9.2, 7.8, 8.8, 6.5],
    'Placed': ['Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)


print("--- First 3 rows of the DataFrame ---")
print(df.head(3))

print("\n--- Summary Information ---")
print(df.info())

print("\n--- Statistical Summary (Numerical Columns) ---")
print(df.describe())
```

```python
# Selecting a single column
names = df['Name']

# Filtering: Extracting all students belonging to the CSE branch
cse_students = df[df['Branch'] == 'CSE']
print("--- CSE Students ---")
print(cse_students)

# Multi-condition filtering: Students with CGPA > 8.0 AND Placed == 'Yes'
high_achievers = df[(df['CGPA'] > 8.0) & (df['Placed'] == 'Yes')]
print("\n--- High Achievers Placed ---")
print(high_achievers)

```

```python
# Grouping by 'Branch' and computing the mean CGPA for each cluster
branch_perf = df.groupby('Branch')['CGPA'].mean().reset_index()
print("--- Average CGPA by Branch ---")
print(branch_perf)

# Multi-level Grouping: Counting placements partitioned across different branches
placement_counts = df.groupby(['Branch', 'Placed']).size().unstack(fill_value=0)
print("\n--- Placement Count Matrix ---")
print(placement_counts)

```

```python
import numpy as np

# Injecting a row with a missing value for demonstration purposes
 df.loc[5] = [106,'Faizan','CSE',np,'No']
# 1. Handling Missing Values: Imputing missing CGPA with the column mean
mean_cgpa = df['CGPA'].mean()
df['CGPA'] = df['CGPA'].fillna(mean_cgpa)

# 2. Creating a calculated column: Converting CGPA to an approximate percentage
df['Percentage'] = df['CGPA'] * 9.5

print("--- Modified and Cleaned DataFrame ---")
print(df)

```
