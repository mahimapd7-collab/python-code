```python
import pandas as pd
df=pd.read_csv('Zomato.csv')
display(df.head())
display(df.tail())
display(df.shape)
display(df.columns)
display(df.info())
display(df.describe())
display(df.isnull().sum())
display(df.duplicated().sum())
display(df.nunique())
display(df.dtypes)
display(df.loc[10:20,["name","online_order"]])
display(df["name"])
display(df.iloc[10:20,0:3])
display(df.dtypes)
# Convert rate to numeric
df["rate"] = pd.to_numeric(df["rate"], errors="coerce")

# Select columns
display(df[["name", "online_order"]])

# Filter data
display(df[(df["listed_in(type)"] == "Dining") & (df["book_table"] == "Yes")])

# Sort by votes
display(df.sort_values(by="votes", ascending=False))

# Rename column
df.rename(columns={"name": "customer_name"}, inplace=True)
display(df.head())


```
